# DATA PRERPOCESSING AND FEATURE ENGINEERING

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
import joblib
```

```python
X = df.drop(columns=['claim','Id','bmi_cat','age_group','region'])
y = df['claim']
print(X.columns)
```

```python
cat_cols = ['gender', 'smoker', 'diabetic']
num_cols = ['age', 'bmi', 'bloodpressure','children']
```

```python
# binary categorical columns
label_encoders = {}
```

```python
for col in cat_cols:
    le = LabelEncoder()
    X[col] = le.fit_transform(X[col])
    label_encoders[col] = le
    
    joblib.dump(le,f'label_encoder_{col}.pkl')
```

```python
label_encoders
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
scaler = StandardScaler()
```

```python
X_train[num_cols] = scaler.fit_transform(X_train[num_cols])
X_test[num_cols] = scaler.transform(X_test[num_cols])
```

```python
joblib.dump(scaler,"scaler.pkl")
```

```python
X_train
```

# Model Selection

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.model_selection import GridSearchCV, cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from xgboost import XGBRegressor
```

```python
def evaluate_model(model,X_train,X_test,y_train,y_test):
    y_pred = model.predict(X_test)
    r2 = r2_score(y_test,y_pred)
    mae = mean_absolute_error(y_test,y_pred)
    rmse = np.sqrt(mean_squared_error(y_test,y_pred))

    return {"R2": r2, "MAE": mae, "RMSE": rmse}
```

```python
results = {}
```

## Linear REgression model

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
results['Linear Regression Model'] = evaluate_model(lr,X_train,X_test,y_train,y_test)
print('Linear Regression Trained')
```

```python
best_poly_model = None
best_poly_score = -np.inf

for degree in [2,3]:
    poly = PolynomialFeatures(degree=degree)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.fit_transform(X_test)

    poly_lr = LinearRegression()
    poly_lr.fit(X_train_poly,y_train)

    score = poly_lr.score(X_test_poly,y_test)

    if score>best_poly_score:
        best_poly_score = score
        best_poly_model = (degree,poly,poly_lr)

degree,poly,poly_lr = best_poly_model

results[f'Polynomial Regression (deg = {degree})'] = evaluate_model(poly_lr,poly.fit_transform(X_train),poly.transform(X_test),y_train,y_test)

print("Polynomial Regression model is trained")
```

## Random Forest Model

```python
rf = RandomForestRegressor()

rf_params = {
    'n_estimators': [100,200],
    'max_depth': [None,10,20],
    'min_samples_split': [2,5],
    'min_samples_leaf': [1,2]
}

rf_grid = GridSearchCV(rf,param_grid=rf_params,cv=5,scoring='r2',n_jobs=-1,verbose=0)
rf_grid.fit(X_train,y_train)
best_rf = rf_grid.best_estimator_

results['Random Forest Model'] = evaluate_model(best_rf,X_train,X_test,y_train,y_test)

print(f" Random Forest model has trained. Best parameters are : {rf_grid.best_params_}")
```

## SVM model

```python
svr = SVR()

svr_param ={
    'kernel': ['rbf','poly','linear'],
    'C': [1,10,50],
    'epsilon':[0.1,0.2,0.5],
    'degree':[2,3]
}

svr_grid = GridSearchCV(svr,param_grid=svr_param,scoring='r2',cv = 5, verbose=0,n_jobs=-1)
svr_grid.fit(X_train,y_train)

best_svr = svr_grid.best_estimator_

results['Support Vector Model'] = evaluate_model(best_svr,X_train,X_test,y_train,y_test)

print(f"SVM model has trained. Best Parameters are : {svr_grid.best_params_}")
```

## XGboost model

```python
xgb = XGBRegressor()

xgb_param = {
    'n_estimators': [100,200],
    'max_depth':[2,3,5,7],
    'learning_rate': [0.01,0.05,0.1],
    'subsample':[0.8,1.0]
}

xgb_grid = GridSearchCV(xgb,param_grid=xgb_param,scoring='r2',n_jobs=-1,verbose=0)
xgb_grid.fit(X_train,y_train)

best_xgb = xgb_grid.best_estimator_

results['XGBoost Model'] = evaluate_model(best_xgb,X_train,X_test,y_train,y_test)

print(f"XGboost model has trained. Best Parameters : {xgb_grid.best_params_}")
```

```python
results
```

```python
results_df = pd.DataFrame(results).T.sort_values(by='R2',ascending=False)
```
